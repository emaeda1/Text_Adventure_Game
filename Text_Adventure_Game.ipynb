```python
#*****************************************************************************
#Assignment Title
#A1: Text Adventure Game Development "Peach Hero Adventure Game"
#*****************************************************************************

#Beginning of Header
#Indicating the date the game was created, the author, format of the file, and 
#source.
"""

Created on October 23rd, 2021

@author: Eri Maeda

File format: Jupyter Notebook(.ipynb format)

Sources:
Pfoundes, C. (1878). Momotaro. The Folk-Lore Record, 1, 123–125.
ASCII Text Art from https://fsymbols.com/text-art/
"""
# End of Header

#DocString describing the background of the game 

#Beginning of DocString
"""
    DocString:

    A)Introduction:

    This game is based on a Japanese folklore called ’Momotaro’ 
    (English title, 'Peach Boy'), which was told since the beginning of 
    the Edo Period (1603-1867).’Momotaro' is a main character and in this game, 
    this fictional hero is recasted from a boy to a girl named ’Peach Girl’ to empower girls. 
    It consists of 3 stages, and has defined functions for start, game clear, 
    and game over. This game requires courage and knowing the story will be helpful. 

        stage1:  Peach Girl is inside Giant Peach. 
        stage2:  Peach Girl is born. Review your weapons to fight Demons.
        stage3a: Bringing your 'kibidango'(Japanese sweet) at your journey.
        stage3b: Gathering friends to fight Demons

    B) Known bug and/or errors: None
    
"""
#End of DocString

#Importing random package
import random 
#Taking input name from the user
name = input(prompt="Enter your name: ")
#Printing welcome message with name which user input with first letter capitalized if it is lower case and the rest lowercase.
print("Hi!", name.capitalize(),","" Welcome to Peach Hero Adventure Game! Let's play the game \|o_o|/")
#Defining weapons;these are the selected weapons at stage2 that are used to fight Demons
weapons=["sword", "swords", "arrow", "arrows", "fireball", "fireballs"]

#Defining a function
def start():
    #Printing the description of the setting for adventurous game to the player.
    print("""
        Once upon a time, there were grandparents living in a peaceful village 
    in Japan. Every day, grandfather went to the mountain to gather firewood and 
    grandmother did laundry by the river.

        One day, a giant peach was floating in the flowing river. 
    It was not like any other day for grandmother. 
    It was a pleasant surprise to see a giant peach.
    She happily said 'It looks so delicious!'
    """)
   
    #Calling stage1 Function
    stage1()
    
#Defining a function   
def stage1():
    #Printing stage1 question and answer choices of 1 to 3. 
    print("""
        [stage1]
    
        This special peach decided to...
        
        1) keep flowing in the river.
        2) be carried out by a friendly grandmother to share the peach with her grandfather.
        3) be carried by a hungry bear who grabbed it quickly.
          
    """)
          
    #For Loop
    #If a user provides an invalid value input, then the player is given another chance up to maximum trials of 5 times.
    #Saving attempts which a user has 
    attempt=5
    #Running while loop as long as user has attempt left 
    while attempt>0:
        #Declaring Variable
        choice=input(prompt="       Answer:")
        #If there is space, removing any spaces or specified characters at the start and end of the string
        choice=choice.strip()
        #If a user insert words in uppercase, it will turn into lowercase.
        choice=choice.lower()
        # Conditional statement(if/elif/else)
        if choice=="1"or "one" in choice or "river" in choice or "flow" in choice:
            # If the player chooses 1, then "Enjoy the river flow and continue to float." is printed.
            print("""
            Enjoy the river flow and continue to float. 
            """)
            #ASCII of text art "Game Over" shows up to inform a user's loss, and the game ends. 
            fail()
            #Stop the loop here because if not, the loop keeps running. 
            break
            
        elif choice=="2" or "two" in choice or "grandmother" in choice or "grandfather" in choice or "friendly" in choice or "share" in choice or "peach" in choice:
            #Proceeding to stage2
            stage2()
            #Stop the loop
            break
                
        elif choice=="3" or "three" in choice or "hungry" in choice or "bear" in choice or "quickly" in choice:
            #ASCII of text art "Game Over" shows up to inform a user the game ended.
            fail()
            #Stop the loop here because if not, the loop keeps running. 
            break
            
        else:
            #If a user types in invalid answer, it will inform a user that it is invalid.
            print(
            """\tInvalid Input:{}\n""".format(choice))
            attempt-=1
            #Provides 5 chances to answer if invalid values are put. 
            #After fifth chance and still fails to answer, the game ends at the game's discretion.
            
#Defining a function
def stage2(): 
    #Printing stage2 question and answers choices of 1 to 3. 
    print("""
        [stage2]
        After grandmother carried a peach and split it in half…
    
        1) You are born a healthy girl and you became a powerful girl.
        2) You are born a diligent girl and you became super good at coding.
        3) You are born but you had poor health and slept in bed.
      """)
    #Declaring Variable (Declaring i as 1 which means the 1st chance. The player is given up to 5 chances to answer.)
    i=1
    #For While Loop (1 if/2 elifs/1 else)
    while i <= 5:
        #Defining Variable
        choice=input(prompt="        Answer:")
        choice=choice.strip()
        choice=choice.lower()
    
        # Conditional Statement (if/elif/else)
        if choice=="1" or "one" in choice or "healthy" in choice or "powerful" in choice:
            #If the player chooses the answer 1, then "You are ready to fight the Demons!" is printed.
            print("""
            You are ready to fight the Demons!
            """)
            #Asking from a user what weapon a user currently has.
            user_weapon=input(prompt="What weapon do you have?\n")
            #Removing the extra space from end and beginning of inputs and converting to lower case
            user_weapon=user_weapon.strip().lower()
            #Checking if the user weapon is a valid weapon to destroy the demon
            if user_weapon in weapons:
                print("\nProceed to stage 3!")
                #Proceeding to stage2.
                stage3()
                break
            else:
                #printing a message to show a user after reviewing the current weapon that a user has
                print("\n\tYou have an invalid weapon. The weapon is not strong enough to destory the demons.\n")
                #Generating a random number from 0,1, and 2. 
                random_number=random.randint(0,2)
                #Using indexing to take out a random weapon name from the list
                weapon_choice=weapons[random_number]
                print("\tAssigning",weapon_choice,"to you!<|0_0|>")
                #Proceeding to stage3
                stage3()
                break
            
            
        elif choice=="2" or "two" in choice or "diligent" in choice or "super" in choice or "coding" in choice:
            #If the player chooses the answer 2, then "You became sick physically and mentally and had to stay in the house" is printed.
            print("\tYou are busy coding and will continue coding.")
            #ASCII of text art "Game Over" shows up to inform the player lost and the game ends.
            fail()
            #Stop the loop here because if not, the loop keeps running.  
            break
            
        elif choice=="3" or "three" in choice or "poor" in choice or "unhealthy" in choice or "bed" in choice or "sleep" in choice:
            #If the player chooses the answer 3, then "You became sick physically and mentally and had to stay in the house" is printed.
            print("""
            You became sick physically and mentally and had to stay in the house.""")
            #ASCII text art of "Game Over" shows up to inform the player lost and the game ends.
            fail()
            #Stop the loop here because if not, the loop keeps running.
            break
            
        else:
            #If a user input invalid answer, then inform a user that it is invalid.
            print("\t Invalid Input:{}\n".format(choice))
        #Loop will continue up to 5 chances.  
        i=i+1 
        
#Defining stage3 (This stage contains stage3a and stage3b)
def stage3():
    #Create a dictionary
    words_num={"zero":0,"one":1,"two":2,"three":3, "four":4, "five":5, "six":6,"seven":7, "eight":8, "nine":9,"ten":10}
    #Printing stage3a question and written-in answer box.
    print("""
        [stage3a]
        You decide to bring delicious kibidangos with you.\n""")
    kibidango=input(prompt="""
How many should you take to the journey to fight the Demons? 
Insert number of kibidango(s) bewteen 1 and 10.\n""")
    kibidango=kibidango.strip()
    #Controlling for Errors (Try/Except)
    #Provding comments to the user about the number of kibidango the user input.
    #Converting type of "kibidango" if the value entered is decimal number. 
    #The float is convered to integer and rounded down. If the number is negative, 
    #then it is converted to positive by taking absolute value. 
    try:
        kibidango=abs(int(float(kibidango))) 
    except:
        #If a user put number in capital letter(s) as words than it will convert it to lowercase 
        kibidango=kibidango.lower()
        #Decoding kibidango inputs (words to numbers)
        kibidango_result=words_num.get(kibidango)
        if not kibidango_result:
            #Printing "That is invalid input. You deserve no kibidango." when a player puts an answer other than numbers.
            print("That is invalid input. You deserve no kibidango.|x_x|")
            kibidango=0 
        else:
            #inputs are alphabetic number than it will be decoded and stored back to kibidango variable
            kibidango=kibidango_result
    #Nested Conditional Statement(2 ifs/1 else)
    #If the kibidago is not equal to zero:
    if kibidango != 0:
        #If number of kibidango is greater than 10, 
        #then printing "You have too many kibidangos. You can only take maximum 10 kibidangos!"
        #Converting the input value to 10 kibidangos.
        if kibidango > 10:
            print("""
            You have too many kibidangos. You can only take maximum 10 kibidangos!""") 
            kibidango = 10
        #If number of kibidango is 10 or less excluding zero, then printing "You get an extra kibidango!"
        #Providing one extra kibidango to the input value(natural number). Not being greedy gives a player, one extra lucky kibidango.
        #No extra kibidango if the value is zero, because the player might be allergic to kibidango. 
        #If the player put negative number, it will take the absolute value of the input number 
        #as the final answer so that the game runs smoothly. 
        else:
            #Printing "You get an extra kibidango!"
            print("\n\tYou get an extra kibidango! |$_$|\n")
            #Adding one extra kibidango to the input value which the player entered.
            kibidango = int(kibidango) + 1 
            print("\tYou have", kibidango, "kibidangos\n")
            
    #Printing the number of kibidango which the player typed in the message.  
    print("\tWe should enjoy our "+str(kibidango)+" kibidango(s) for a long journey to come.\n")
    
    #Printing the question to stage3b.
    print("""
        [stage3b]
        Choose what you want to do.
    
        1) You meet your friends, dog, monkey, and pheasant. Together, you go to fight the Demons.
        2) You meet your friends, dog and monkey. Decide to feed them with "kibidango"and live together.
        3) You decide to fight the Demons alone.
    """)
    #For Loop 
    for i in [0,1,2,3,4]:
        choice=input(prompt="        Answer:")
        choice=choice.strip()
        choice=choice.lower()
        if choice=="1" or "one" in choice or "pheasant" in choice or "together" in choice:
            #If the player chooses the answer 1, then"Your friends will be a great support to fight the demons." is printed.
            print("""
            Your friends will be a great support to fight the Demons.
            """)
            #Proceeding to gameclear()
            gameclear()
            #Stop the loop here, because if not, the loop keeps going.
            break
        elif choice=="2" or "two" in choice or "dog and monkey" in choice or "feed" in choice or "kibidango" in choice:
            #If a user chooses the answer 2, then "'Kibidango' is the best Japanese sweet to eat!" is printed.
            print("""
            'Kibidango'is the best Japanese sweet to eat!\n""")
            #ASCII of text art "Game Over" shows up to inform the player lost and the game ends.
            fail()
            #Stop the loop here because if not, the loop keeps running.
            break
        elif choice=="3" or "three" in choice or "alone" in choice:
            #If a user answers 3, then fighting takes place. 
            print("\tDemons are still alive and continue to destroy the village and have a fight!!! \n ATTACKKKKK!!!\n")
            #A game will randomly decides whether a user wins a game or looses a game. 
            demon = random.randint(0,2)
            #If a user looses, printing"You aree defeated as you cannot over power demons."
            if demon == 0:
                print("\tYou are defeated as you cannot overpower demons.\n")
                fail()
            #If a usr wins, printing"You WIN!!!")
            else: 
                print("\tYou WIN!!!")
                gameclear()
            
            #ASCII text art of "Game Over" shows up to inform the player lost and the game ends.

            #Stop the loop here because if not, the loop keeps running.
                break
        else:
                #Printing Invalid Input which the player put. 
            print("\t Invalid Input:{}\n".format(choice))
            
#Defining a function
def fail():
    #Printing "Demons are still alive and continue to destroy the village".
    print("""
            Demons are still alive and continue to destroy the village.
    """)
    #Printing ASCII text art of "Game Over"
    print("""
    
  ░██████╗░░█████╗░███╗░░░███╗███████╗  ░█████╗░██╗░░░██╗███████╗██████╗░
  ██╔════╝░██╔══██╗████╗░████║██╔════╝  ██╔══██╗██║░░░██║██╔════╝██╔══██╗
  ██║░░██╗░███████║██╔████╔██║█████╗░░  ██║░░██║╚██╗░██╔╝█████╗░░██████╔╝
  ██║░░╚██╗██╔══██║██║╚██╔╝██║██╔══╝░░  ██║░░██║░╚████╔╝░██╔══╝░░██╔══██╗
  ╚██████╔╝██║░░██║██║░╚═╝░██║███████╗  ╚█████╔╝░░╚██╔╝░░███████╗██║░░██║
  ░╚═════╝░╚═╝░░╚═╝╚═╝░░░░░╚═╝╚══════╝  ░╚════╝░░░░╚═╝░░░╚══════╝╚═╝░░╚═╝
  """)
        

#Defining a function
def gameclear():
    #Printing "You won!Congratulations. Please is back in the village".
    print("""
        You won! CONGRATULATIONS!! Peace is back in the village.""")
    #Printing ASCII text art of "Game Clear"
    print("""
    
  ░██████╗░░█████╗░███╗░░░███╗███████╗  ░█████╗░██╗░░░░░███████╗░█████╗░██████╗░
  ██╔════╝░██╔══██╗████╗░████║██╔════╝  ██╔══██╗██║░░░░░██╔════╝██╔══██╗██╔══██╗
  ██║░░██╗░███████║██╔████╔██║█████╗░░  ██║░░╚═╝██║░░░░░█████╗░░███████║██████╔╝
  ██║░░╚██╗██╔══██║██║╚██╔╝██║██╔══╝░░  ██║░░██╗██║░░░░░██╔══╝░░██╔══██║██╔══██╗
  ╚██████╔╝██║░░██║██║░╚═╝░██║███████╗  ╚█████╔╝███████╗███████╗██║░░██║██║░░██║
  ░╚═════╝░╚═╝░░╚═╝╚═╝░░░░░╚═╝╚══════╝  ░╚════╝░╚══════╝╚══════╝╚═╝░░╚═╝╚═╝░░╚═╝
  """)

    
#Start the game
start()
```
